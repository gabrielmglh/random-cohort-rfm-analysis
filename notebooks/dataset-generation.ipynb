## Generation of random data for analysis 

In this notebook, I generate the random data necessary for our analysis. The analysis is generated using parameters that turns the data more realistic such as customers that buy many times overtime. The period that I've chosen is of two years. 

```python
import numpy as np
import pandas as pd 

# parameters
num_customers = 1000
max_orders_per_customer = 15 # we define this just for simplicity

# date range for order dates
start_date = pd.to_datetime('2020-01-01')
end_date = pd.to_datetime('2022-01-01')

# list where the transaction data will be stored
transactions = []

# counter to keep track of order IDs
order_id_counter = 1

# loop for each customer
for customer_id in range(1, num_customers + 1):
    # generate random date for the first order
    first_purchase = start_date + pd.to_timedelta(np.random.randint(0, (end_date - start_date).days), unit='D')
    # random number of orders for this customer
    num_orders = np.random.randint(1, max_orders_per_customer + 1)
    # loop for each order of the customer
    for order_num in range(num_orders):
        # generate random date for the order, ensuring it's after the first purchase
        order_date = first_purchase + pd.to_timedelta(np.random.randint(0, 365), unit='D')
        # generate random order amount between R$10 and R$500
        order_amount = round(np.random.uniform(10, 500), 2)
        # add the order to the dataset
        transactions.append((customer_id, order_id_counter, order_date.date(), order_amount))
        # increment the order ID counter
        order_id_counter += 1 

# create a dataframe from the transactions list
df_transactions = pd.DataFrame(transactions, columns=['customer_id', 'order_id', 'order_date', 'order_amount'])

# order the dataset by date
df_transactions.sort_values(by='customer_id', inplace=True)
```
